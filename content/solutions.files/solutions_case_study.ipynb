```python
import pandas as pd
import seaborn as sns
%matplotlib inline
sns.set_theme()

df = pd.read_csv("../data/Library_Usage_Clean.csv",
                low_memory=False
                )
sns.displot(df['Total Checkouts'], kde=False)
```

```python
df_filtered = df[df['Total Checkouts'] < 800]
sns.displot(df_filtered['Total Checkouts'], kde=False)
```

```python
df['Total Checkouts Sqrt'] = df['Total Checkouts']**0.5
sns.displot(df['Total Checkouts Sqrt'], kde=False)
```

```python
import numpy as np
df['Checkouts Per Month'] = df['Total Checkouts']/df['Membership Duration']
df['Checkouts Per Month'].replace(np.inf, np.nan, inplace=True)
df['Checkouts Per Month'].fillna(0, inplace=True)
```

```python
sns.distplot(df['Checkouts Per Month']**0.5, kde=False)
```
